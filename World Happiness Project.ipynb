```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR=DecisionTreeRegressor()
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state =i)
    mod = DecisionTreeRegressor()
    mod.fit(x_train,y_train)
    pred = mod.predict(x_test)
    acc=r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best r2_score is ",maxAccu," on Random_state ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=43)
```

# Modeling

```python
#importing necessary libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

# Random Forest Regressor

```python
RT=RandomForestRegressor()
RT.fit(x_train,y_train)
pred=RT.predict(x_test)
r2=r2_score(y_test,pred)
print(r2)
```

# SVM Regressor

```python
svr=SVR()
svr.fit(x_train,y_train)
pred=svr.predict(x_test)
r2=r2_score(y_test,pred)
print(r2)
```

# KNN Regressor

```python
knn=KNN()
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
r2=r2_score(y_test,pred)
print(r2)
```

# Cross Validation

```python
print(cross_val_score(RT,x,y,cv=5).mean())
```

```python
print(cross_val_score(DTR,x,y,cv=5).mean())
```

```python
print(cross_val_score(svr,x,y,cv=5).mean())
```

```python
print(cross_val_score(knn,x,y,cv=5).mean())
```

The least in the difference is our best model, so here DecisionTreeClassifier is our best model
